```python
### Packages
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine.data import mtcars
```

```python
### Example
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
+ geom_point()
+ stat_smooth(method='lm'))
```

```python
### Example
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
+ geom_point()
+ stat_smooth(method='lm')
+ facet_wrap('~gear'))
```

```python
### Packages
import pandas as pd
import numpy as np
from plotnine import *
```

```python
### Data
from plotnine.data import mpg
```

```python
### Data
data_raw = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/ggplot2/mpg.csv")
```

- `manufacturer`: 자동차 제조사
- `displ`: 자동차 배기량
- `cyl`: 실린더 수
- `drv`: 구동 방식
- `cty`: 도심 연비
- `hwy`: 고속도로 연비
- `fl`: 연료 종류

```python
data_raw.shape
```

```python
data_raw.head()
```

```python
data_raw.describe(include = 'all')
```

```python
### Remove first column
data_use = data_raw.copy()
data_use = data_use.iloc[:,1:]
data_use
```

```python
### ggplot: 시각화 자료 + 시각화 특성
ggplot(data_use, aes(x = 'displ', y = 'hwy'))
```

```python
### ggplot: 시각화 자료 + 시각화 특성 + 시각화 종류
ggplot(data_use, aes(x = 'displ', y = 'hwy')) + geom_point()
```

```python
### Smooth line & SE
ggplot(data_use, aes(x = 'displ', y = 'hwy')) + geom_smooth()
```

```python
### Point + Smooth line
ggplot(data_use, aes(x = 'displ', y = 'hwy')) + geom_smooth() + geom_point()
```
